```python
import numpy as np
import matplotlib.pyplot as plt

#Paso 1: Generamos puntos aleatorios
num_galaxias = 10000
min_x, max_x = 0, 10000
min_y, max_y = 0, 10000


#Generamos coordenadas x e y aleatorias para las galaxias
x_galaxias = np.random.uniform(min_x, max_x, num_galaxias)
y_galaxias = np.random.uniform(min_y, max_y, num_galaxias)

plt.figure(figsize=(8, 6))
plt.scatter(x_galaxias, y_galaxias, color='blue', s=10)
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Distribucin de Galaxias')
plt.grid(True)
plt.show()



#Paso 2: Definimos los intervalos de distancia
num_bins = 100
distancia_max = np.sqrt((max_x - min_x)**2 + (max_y - min_y)**2)
bins = np.linspace(0, distancia_max, num_bins + 1)




```

```python
#Paso 3: Contamos los pares de puntos
RR = np.zeros(num_bins)

for i in range(num_galaxias):
    for j in range(i + 1, num_galaxias):
        distancia = np.sqrt((x_galaxias[i] - x_galaxias[j])**2 + (y_galaxias[i] - y_galaxias[j])**2)
        bin_index = int(distancia / (distancia_max / num_bins))
        RR[bin_index] += 1
```

### Distribucin de galaxias en una dona

```python
import numpy as np
import matplotlib.pyplot as plt

#Paso 1: Generamos puntos aleatorios en coordenadas polares
box_size=10000
num_galaxias = 10000
min_r, max_r = 2900, 3000  #Radio interior y exterior de la dona
min_theta, max_theta = 0, 2 * np.pi

#Generamos radios y ngulos aleatorios para las galaxias en coordenadas polares
radios=np.random.uniform(min_r, max_r, num_galaxias)
angulos=np.random.uniform(min_theta, max_theta, num_galaxias)

#Convertimos las coordenadas polares a coordenadas cartesianas
x_dona=radios*np.cos(angulos)+box_size/2
y_dona=radios*np.sin(angulos)+box_size/2

#Graficamos los puntos en el plano cartesiano
plt.figure()
plt.scatter(x_dona, y_dona, color='blue', s=10)
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
plt.title("Distribucin de Galaxias en una Dona (Coordenadas Cartesianas)")
plt.xlim(0, 10000)
plt.grid(True)
plt.show()

```

```python
#Definimos los intervalos de distancia
num_bins=100



#Contamos los pares de puntos
DD_dona= np.zeros(num_bins)

for i in range(num_galaxias):
    for j in range(i + 1, num_galaxias):
        #Calculamos la distancia euclidiana entre dos puntos en la dona
        distancia = np.sqrt((x_dona[i] - x_dona[j])**2 + (y_dona[i] - y_dona[j])**2)
        #Calculamos el ndice del bin utilizando la funcin de parte entera
        bin_index2 = int(distancia / (distancia_max/num_bins))
        #Verificamos si el ndice est dentro de los lmites
        if bin_index2 < num_bins:
            #Contamos los pares de puntos en cada intervalo
            DD_dona[bin_index2] += 1

```

```python
#Estimador Peebles & Hauser
Xi_dona=(DD_dona/RR)-1

plt.plot(bins[0:-1], Xi_dona)
plt.xlabel("Distancia")
plt.ylabel("Correlacin Dona")
plt.title("Funcin de correlacin de galaxias Peebles & Hauser")
plt.show()
```

### Anillos  

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el nmero de anillos y el nmero de puntos en cada anillo
num_anillos = 10
num_puntos_por_anillo = 1000
radio = 2000  # Radio de los anillos

# Generamos centros aleatorios de los anillos dentro de la caja
centros_x = np.random.uniform(radio, 10000 - radio, num_anillos)
centros_y = np.random.uniform(radio, 10000 - radio, num_anillos)

# Listas para almacenar las coordenadas de los puntos
x_galaxy = []
y_galaxy = []

# Generamos los puntos para cada anillo
for i in range(num_anillos):
    centro_x = centros_x[i]
    centro_y = centros_y[i]

    # Generamos ngulos equidistantes alrededor del centro
    angulos = np.linspace(0, 2*np.pi, num_puntos_por_anillo)

    # Calculamos las coordenadas de los puntos
    x_anillo = centro_x + np.cos(angulos) * radio
    y_anillo = centro_y + np.sin(angulos) * radio

    # Agregamos las coordenadas a las listas
    x_galaxy.extend(x_anillo)
    y_galaxy.extend(y_anillo)

# Grfica
plt.figure(figsize=(8, 8))
plt.scatter(x_galaxy, y_galaxy, color='blue', s=10)
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Distribucin de Galaxias: 10 Anillos con 1000 Puntos cada uno')
plt.xlim(-1000, 11000)
plt.ylim(-1000, 11000)
plt.grid(True)
plt.show()
```
